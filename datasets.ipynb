```python
import os
import json
import csv
import re

def extract_ingredients():
    """Extrae ingredientes únicos de archivos JSON, CSV, MD y TXT"""
    ingredients = set()
    
    def walk(obj):
        """Recorre recursivamente estructuras anidadas buscando 'ingredients'"""
        if isinstance(obj, dict):
            for k, v in obj.items():
                if k.lower() == 'ingredients':
                    if isinstance(v, (list, tuple)):
                        for it in v:
                            # Maneja listas anidadas
                            if isinstance(it, (list, tuple)):
                                ingredients.update(str(x).strip() for x in it if x)
                            else:
                                ing_str = str(it).strip().strip("'\"[]")
                                if ing_str:
                                    ingredients.add(ing_str)
                    elif v:
                        ing_str = str(v).strip().strip("'\"[]")
                        if ing_str:
                            ingredients.add(ing_str)
                else:
                    walk(v)
        elif isinstance(obj, list):
            for el in obj:
                walk(el)
    
    # Procesar archivos
    for root, _, files in os.walk('.'):
        for f in files:
            p = os.path.join(root, f)
            try:
                if f.endswith('.json'):
                    with open(p, 'r', encoding='utf-8') as fh:
                        walk(json.load(fh))
                elif f.endswith('.csv'):
                    with open(p, newline='', encoding='utf-8') as fh:
                        for row in csv.DictReader(fh):
                            for k, v in row.items():
                                if k and 'ingredient' in k.lower() and v:
                                    for it in re.split(r'[;,\[\]"\']', v):
                                        ing_str = it.strip()
                                        if ing_str:
                                            ingredients.add(ing_str)
                elif f.endswith(('.md', '.txt')):
                    with open(p, encoding='utf-8') as fh:
                        for line in fh:
                            if 'ingredient' in line.lower():
                                for part in re.split(r'[:,\[\]"\']', line):
                                    ing_str = part.strip()
                                    if ing_str:
                                        ingredients.add(ing_str)
            except Exception as e:
                print(f"Error procesando {p}: {e}")
    
    return sorted(ingredients)

# Ejecutar
ingredientes = extract_ingredients()
print(f"Total de ingredientes únicos: {len(ingredientes)}\n")
for ing in ingredientes:
    print(ing)
```
