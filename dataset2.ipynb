```python
import pandas as pd

df = pd.read_csv('develop/config/dishes_data_MAPEADO.csv')
```

```python
from collections import Counter
import re

# Contar todos los ingredientes
todos_ingredientes = []

for ing_str in df['ingredients'].dropna():
    for part in re.split(r'[,]', ing_str):
        ing_limpio = part.strip()
        if ing_limpio:
            todos_ingredientes.append(ing_limpio)

# Contar ocurrencias
contador = Counter(todos_ingredientes)

# Mostrar todos los ingredientes con sus counts
print("Ingredientes con sus counts:\n")
print(f"{'Ingrediente':<40} {'Count':<10}")
print("-" * 50)

for ing, count in contador.most_common():
    print(f"{ing:<40} {count:<10}")

print(f"\n✓ Total de ingredientes únicos: {len(contador)}")
print(f"✓ Total de ocurrencias: {sum(contador.values())}")
```

```python
# Aplicar el mapping al dataset
def mapear_ingredientes(ingredientes_str):
    if pd.isna(ingredientes_str):
        return ingredientes_str
    
    ingredientes_procesados = []
    
    for part in re.split(r'[,]', ingredientes_str):
        ing_str = part.strip().lower()
        if ing_str:
            # Buscar en el mapping
            if ing_str in mapping:
                mapped = mapping[ing_str]
                if mapped != "REMOVE":  # No agregar ingredientes marcados como REMOVE
                    ingredientes_procesados.append(mapped)
            else:
                # Si no está en el mapping, mantener el ingrediente original
                ingredientes_procesados.append(ing_str)
    
    return ", ".join(ingredientes_procesados) if ingredientes_procesados else ""

# Aplicar la función al dataset
df['ingredients'] = df['ingredients'].apply(mapear_ingredientes)

# Guardar el resultado
df.to_csv('develop/config/dishes_data_MAPEADO_FINAL.csv', index=False)
print("✓ Archivo guardado como: develop/config/dishes_data_MAPEADO_FINAL.csv")

# Mostrar primeros registros
print(f"\nPrimeros 5 registros:\n")
print(df[['name', 'ingredients']].head())

# Contar ingredientes finales
todos_ingredientes = []
for ing_str in df['ingredients'].dropna():
    for part in re.split(r'[,]', ing_str):
        ing_limpio = part.strip()
        if ing_limpio:
            todos_ingredientes.append(ing_limpio)

contador = Counter(todos_ingredientes)
print(f"\n✓ Total de ingredientes únicos: {len(contador)}")
print(f"✓ Total de ocurrencias: {sum(contador.values())}")
```
